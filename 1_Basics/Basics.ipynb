## Matplotlib Intro

Activity 1: Matplotlib Intro

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x, y)
plt.show()
```

```python
#Plotting the Job Postings Date

date_counts = df.job_posted_date.dt.date.value_counts().sort_index()
date_counts
```

```python
plt.plot(date_counts.index,date_counts)
```

```python
month_counts = df.job_posted_date.dt.month.value_counts().sort_index()
plt.plot(month_counts.index,month_counts.values)
plt.show()
```

```python
#Number of Job Posted by Job Title

job_title_counts = df.job_title_short.value_counts().sort_values()

plt.barh(y=job_title_counts.index,width=job_title_counts.values)
plt.show()
```

Activity 2: Matplotlib labeling

```python
job_counts = df.job_title_short.value_counts().sort_values()

plt.barh(y=job_counts.index,width=job_counts.values)
plt.xlabel('Number of Job Postings')
plt.ylabel('Job Title')
plt.title('Number of Job Postings by Job Title')
plt.show()
```

```python
plt.bar(x=job_counts.index,height=job_counts.values)
plt.xlabel('job title')
plt.ylabel('numbers of job posted')
plt.title('Number of Job Posted per Title')
plt.xticks(rotation = 45, ha='right')

#To show values above the bar
for i, v in enumerate(job_counts.values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom')

plt.show()
```

Activity 2: Basic Matplotlib & Pandas

```python
job_counts.sort_values(ascending=False).plot(kind = 'bar')
plt.xlabel('')
plt.ylabel('counts of posting')
plt.title('Number of Posted Job')
plt.xticks(rotation = 45, ha='right')
plt.show()
```
