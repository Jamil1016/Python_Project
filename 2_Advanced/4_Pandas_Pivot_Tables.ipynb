```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df.job_posted_date)

df.info()
```

```python
df.pivot_table(values='salary_year_avg',index='job_country',columns='job_title_short',aggfunc='median')
```

```python
top_countries = df.job_country.value_counts().head(6).index
top_countries
```

```python
df_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'
    )

df_country_salary = df_country_salary.loc[top_countries]

df_country_salary
```

```python
job_titles = ['Data Analyst','Data Engineer','Data Scientist']

df_country_salary = df_country_salary.loc[:,job_titles]
df_country_salary
```

```python
df_country_salary.plot(kind='bar')
plt.xlabel('')
plt.ylabel('Median Salary ($USD)')
plt.title('Median Salary by Countrt and Job Title')
plt.show()
```
